## 2. Conditional Probability
Given the following table of events, compute P(A∣B) and P(B∣A)
|           |     B       |   not B         |
|-----------|-------------|-----------------|
|   A       |   30        |      20         |
|   not A   |   50        |     100         |



```python
na=30+20
nb=30+50
n_a_and_b=30
total=30+20+50+100

pa=na/total
pb=nb/total
p_a_and_b=n_a_and_b/total

p_a_b=p_a_and_b/pb
p_b_a=p_a_and_b/pa
print(f"P(A|B)={p_a_b}\nP(B|A)={p_b_a}")
```
